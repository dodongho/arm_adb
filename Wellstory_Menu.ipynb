## (실습) 대표 메뉴만 보이기

````
import requests
import pprint

request_url = "http://www.samsungwelstory.com/menu/getMenuList.do"
request_data = {          
    "restaurant_code": "",          
    "dt": "20190527",          
    "oFlag": "R",          
    "sFlag": "research",          
    "hallFlag": "",          
    "sDate": "2019-05-27",          
    "hall_no": "E59U,E59T,E59S,E59W,E59V,E59Z,E5A1,E5A4,E5A5,E59X,E59Y,E5A2,E5A3,E5A6"     
}

res = requests.post(request_url, data=request_data)

result = []
if 200 == res.status_code:
    data = res.json()
    
    for d in data:
      if "Y"== d["first_row_yn"]:
        result.append(d["menu_name"])
    pprint.pprint(result)
    print("Total data : ", len(result))
else:
    print("Getting data failed as %d" % res.status_code)
````

## (실습) 점심 대표 메뉴 + total kcal 표시하기

````
import requests
import pprint

request_url = "http://www.samsungwelstory.com/menu/getMenuList.do"
request_data = {          
    "restaurant_code": "",          
    "dt": "20190527",          
    "oFlag": "R",          
    "sFlag": "research",          
    "hallFlag": "",          
    "sDate": "2019-05-27",          
    "hall_no": "E59U,E59T,E59S,E59W,E59V,E59Z,E5A1,E5A4,E5A5,E59X,E59Y,E5A2,E5A3,E5A6"     
}

res = requests.post(request_url, data=request_data)

result = []
if 200 == res.status_code:
    data = res.json()
    
    for d in data:
      if "Y"== d["first_row_yn"] and "2" == d["menu_meal_type"]:
        result.append((d["menu_name"], d["tot_kcal"]))
    pprint.pprint(result)
    print("Total data : ", len(result))
else:
    print("Getting data failed as %d" % res.status_code)
````

## (실습) 아침/점심/저녁 대표 메뉴를 제공 식당 별로 data 재구성해 보여주기

````
import requests
import pprint

request_url = "http://www.samsungwelstory.com/menu/getMenuList.do"
request_data = {          
    "restaurant_code": "",          
    "dt": "20190527",          
    "oFlag": "R",          
    "sFlag": "research",          
    "hallFlag": "",          
    "sDate": "2019-05-27",          
    "hall_no": "E59U,E59T,E59S,E59W,E59V,E59Z,E5A1,E5A4,E5A5,E59X,E59Y,E5A2,E5A3,E5A6"     
}

meal_type = {
    "1": "아침", 
    "2": "점심", 
    "3": "저녁", 
    "4": "야식"
}

result = {
    "아침": [],
    "점심": [],
    "저녁": [],
    "야식": []
}

res = requests.post(request_url, data=request_data)

if 200 == res.status_code:
    data = res.json()
    
    """
    "아침": [
        ("식당이름1", "대표메뉴1", "총칼로리1"),
        ("식당이름2", "대표메뉴2", "총칼로리2"),
        ...,
        ...
    ]
    """
    
    for d in data:
      if "Y"== d["first_row_yn"]:
        m_type = meal_type[d["menu_meal_type"]]
        result[m_type].append((d["course_txt"], d["menu_name"], d["tot_kcal"]))
               
    pprint.pprint(result)
    print("Total data : ", len(result))
else:
    print("Getting data failed as %d" % res.status_code)
````

````
import requests
import pprint

request_url = "http://www.samsungwelstory.com/menu/getMenuList.do"
request_data = {          
    "restaurant_code": "",          
    "dt": "20190527",          
    "oFlag": "R",          
    "sFlag": "research",          
    "hallFlag": "",          
    "sDate": "2019-05-27",          
    "hall_no": "E59U,E59T,E59S,E59W,E59V,E59Z,E5A1,E5A4,E5A5,E59X,E59Y,E5A2,E5A3,E5A6"     
}

meal_type = {
    "1": "아침", 
    "2": "점심", 
    "3": "저녁", 
    "4": "야식"
}

result = {
    "아침": {},
    "점심": {},
    "저녁": {},
    "야식": {}
}

res = requests.post(request_url, data=request_data)

if 200 == res.status_code:
    data = res.json()
    
    """
    "아침": {
        "식당이름1": {
            "menu_name": string,
            "tot_kcal": string,
            "sub_menu": list
        },
        "식당이름2": {
            "menu_name": string,
            "tot_kcal": string,
            "sub_menu": list
        },
        ...
        ...
    }
    """
    
    for d in data:
      m_type = meal_type[d["menu_meal_type"]]
      restraunts = list(result[m_type].keys())
      restraunt = d["course_txt"]
      
      if restraunt not in restraunts:
        result[m_type][restraunt] = {
            "menu_name": "",
            "tot_kcal": "",
            "sub_menu": []
        }
        
      if "Y"== d["first_row_yn"]:
        result[m_type][restraunt] .update(
            {
                "menu_name": d["menu_name"],
                "tot_kcal": d["tot_kcal"]
            }
        )
      else:
        result[m_type][restraunt]["sub_menu"].append(d["menu_name"])
               
    pprint.pprint(result)
    print("Total data : ", len(result))
else:
    print("Getting data failed as %d" % res.status_code)
````
